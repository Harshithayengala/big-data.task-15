```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("BigDataBasics") \
    .getOrCreate()

data = [("Alice", 34), ("Bob", 28), ("Charlie", 45), ("Diana", 31)]
columns = ["Name", "Age"]

df = spark.createDataFrame(data, columns)
df.show()
```
